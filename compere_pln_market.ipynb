```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from utils import annullised_mean, annullised_std

plt.rcParams["figure.figsize"] = (16, 8)
```

Import data


```python
acwi_euro = pd.read_excel('oryginal_price_changes.xlsx', usecols=['Date', 'ACWI'])
acwi_euro.set_index('Date', inplace=True)
```

```python
assets_pln = pd.read_excel('pln_price_changes.xlsx')
assets_pln.set_index('Date', inplace=True)
```

Compere 36 months annulised average ACWI EUR and ACWI PLN 

```python
acwi_euro_return = acwi_euro[['ACWI']].rolling(36).apply(annullised_mean)
acwi_pln_return = assets_pln[['ACWI']].rolling(36).apply(annullised_mean)
```

```python
fig, ax = plt.subplots()
sns.lineplot(data=acwi_euro_return, ax=ax)
sns.lineplot(data=acwi_pln_return, ax=ax)
ax.legend(['ACWI_EUR', 'ACWI_PLN'])
```

```python
euro_return = assets_pln['EURPLN'].rolling(36).apply(annullised_mean)
usd_return = assets_pln['USDPLN'].rolling(36).apply(annullised_mean)
```

ACWI 36 month moving annulised avering in pln compared to pln and usd

```python
fig, ax = plt.subplots()
sns.lineplot(data=acwi_euro_return, ax=ax)
sns.lineplot(data=euro_return, ax=ax)
sns.lineplot(data=usd_return, ax=ax)
ax.legend(['ACWI', 'EUR', 'USD'])
```

Compere MWIG40 to acwi, developed

```python
mov_avgs = assets_pln[['ACWI', 'World', 'Mwig40']].rolling(36).apply(annullised_mean)
```

```python
sns.lineplot(data=mov_avgs)
```

Cumulative return

```python
cum_return = (1+assets_pln[['ACWI', 'World', 'Mwig40']]).cumprod()-1
sns.lineplot(data=cum_return)
```

```python
mov_std = assets_pln[['ACWI', 'World', 'Mwig40']].rolling(36).apply(annullised_std)
sns.lineplot(data=mov_std)
```

36 months moving corelation

```python
moving_corr = assets_pln['World'].rolling(36).corr(assets_pln['Mwig40'])
sns.lineplot(data=moving_corr, ax=ax, palette='mako')
```
