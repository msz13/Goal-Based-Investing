```python
import pandas as pd
import pandas_datareader.data as web
import datetime
import statsmodels.api as sm
from statsmodels.tsa.api import VAR
from statsmodels.tsa.vector_ar.var_model import VARProcess
import statsmodels.api as sm
import numpy as np
from preproccessing import loadStooqData, loadYahooData
import matplotlib.pyplot as plt
import seaborn as sns
import preproccessing as prep
```

```python
cpipl = prep.loadStooqData('cpiypl.m','2003-01-01',frequency='m')
cpipl.name = 'cpipln'
cpius = prep.loadStooqData('cpiyus.m','2003-01-01',frequency='m')
cpius.name = 'cpius'
usdpln = prep.loadStooqData('cpiyus.m','2003-01-01',frequency='m')
usdpln.name = 'usdpln'
```

```python
fig, ax = plt.subplots(figsize=(20,6))

cpipl.plot(ax=ax, color='blue', legend=True)
cpius.plot(ax=ax, color='orange', legend=True)

ax2 = ax.twinx()
usdpln.plot(ax=ax2, color='green', legend=True)
```

```python
ppp = np.log((cpipl/100+1).cumprod()) - np.log((cpius/100+1).cumprod()) 

fig, ax = plt.subplots(figsize=(20,6))
ppp.plot(ax=ax, color='red')

ax2 = ax.twinx()
usdpln.plot(ax=ax2, color='green', legend=True)
```
